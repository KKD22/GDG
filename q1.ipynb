```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import yfinance as yf
import io
from io import StringIO
import csv  
```

```python
page = requests.get(url="https://www.screener.in/company/NIFTY/?order=desc&sort=return+over+1year&limit=50&page=1")
soup = BeautifulSoup(page.text, 'html.parser')
#<table class="data-table text-nowrap striped mark-visited">
table = soup.find('table')
titles = table.find_all('th')
all_title = [title.text.strip() for title in titles]
print(all_title)
```

```python
df1 = pd.DataFrame(columns= all_title[:-12])
df1
```

```python
col_data = soup.find_all('tr')
```

```python
col_data_1 = col_data[1:16]
```

```python
for row in col_data_1:
    row_data_1 = row.find_all('td')
    ind_data_1 = [data1.text.strip() for data1 in row_data_1]
    length = len(df1)
    df1.loc[length] = ind_data_1
```

```python
df2 = pd.DataFrame(columns= all_title[:-12])
df2
```

```python
col_data_2 = col_data[17:-1]
```

```python
for row2 in col_data_2:
    row_data_2 = row2.find_all('td')
    ind_data_2 = [data2.text.strip() for data2 in row_data_2]
    length = len(df2)
    df2.loc[length] = ind_data_2
```

```python
df_merged = pd.concat([df1, df2], ignore_index=True)

```
